```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "/mnt/data/DATA-CODELEB-2-MD-2.csv"
df = pd.read_csv("DATA-CODELEB-2-MD-2.csv")
```

```python
display(df.head(7))
```

```python
df.info()
```

```python
# Konversi kolom 'Dues amount' menjadi numerik (hapus simbol jika ada)
df['Dues amount'] = df['Dues amount'].replace('[^0-9.]', '', regex=True)
df['Dues amount'] = pd.to_numeric(df['Dues amount'], errors='coerce')
```

```python
# 3. Hitung total dan rata-rata kontribusi
total_kontribusi = df['Dues amount'].sum()
rata_rata_kontribusi = df['Dues amount'].mean()

print("\nTotal Kontribusi:", total_kontribusi)
print("Rata-rata Kontribusi:", rata_rata_kontribusi)
```

```python
# 4. Visualisasi 1: Histogram Distribusi Kontribusi
plt.figure()
plt.hist(df['Dues amount'].dropna())
plt.title("Distribusi Kontribusi (Dues Amount)")
plt.xlabel("Kontribusi")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# 4. Visualisasi 2: Bar Chart Kontribusi per Member
plt.figure()
plt.bar(df['Member ID'], df['Dues amount'])
plt.title("Kontribusi per Member")
plt.xlabel("Member ID")
plt.ylabel("Kontribusi")
plt.xticks(rotation=90)
plt.show()
```
